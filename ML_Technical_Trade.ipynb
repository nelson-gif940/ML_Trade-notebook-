```python
import numpy as np
import os
import sys
import shutil
import matplotlib.pyplot as plt
```

#### NECESSARY FUNCTIONS

```python
def splt(string):
    output = string.split("_")
    output_gif=output[4].split(".")
    output_gif=output_gif[0]+"."+output_gif[1]
    output=[float(output[1]),float(output[2]),float(output[3]),float(output_gif)] #low,high,open,close
    return output
```

### 1. CLASSIFICATION

#### NECESSARY LIBRAIRIES

```python
import tensorflow as tf
import pathlib

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
```

#### 1.1 Preparing the Data -------------------------------------------------------------------

```python
data_dir = pathlib.Path('C:/Users/nelso/Desktop/ML_Classification/Sorted/') #Path with the Sorted Folder

#Batch proprieties 

batch_size = 32
img_height = 180
img_width = 180

#Training Set (80% - 20% Split)

train_ds = tf.keras.utils.image_dataset_from_directory(
  data_dir,
  validation_split=0.2,
  subset="training",
  seed=125,
  image_size=(img_height, img_width),
  batch_size=batch_size)

#Validation Set (80% - 20% Split)

val_ds = tf.keras.utils.image_dataset_from_directory(
  data_dir,
  validation_split=0.2,
  subset="validation",
  seed=125,
  image_size=(img_height, img_width),
  batch_size=batch_size)

#Grayscale Batch transformation + Normalization

normalization_layer = layers.Rescaling(1./255)

normalized_ds = train_ds.map(lambda x, y: (normalization_layer(x), y))
image_batch, labels_batch = next(iter(normalized_ds))
first_image = image_batch[0]
print(np.min(first_image), np.max(first_image))

class_names = train_ds.class_names
print(class_names)
num_classes = len(class_names)
```

#### 1.2 Training -------------------------------------------------------------------

Training the classification

```python
model = Sequential([
  layers.Rescaling(1./255, input_shape=(img_height, img_width, 3)),
  layers.Conv2D(16, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(32, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(64, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Flatten(),
  layers.Dense(128, activation='relu'),
  layers.Dense(num_classes)
])

model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
epochs = 10 #number of run
history = model.fit(
  train_ds,
  validation_data=val_ds,
  epochs=epochs
)
```

#### 1.3 Visualisation of the Result ------------------------------------------

```python
import matplotlib.pyplot as plt

acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(epochs)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

### 2. Testing

#### NECESSARY LIBRAIRIES

```python
import tensorflow as tf
import pathlib

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential

import numpy as np
import os
import sys
import shutil
import matplotlib.pyplot as plt
from IPython.display import Image
```

#### 2.0 Loading the Model

```python
model = keras.models.load_model('C:/Users/nelso/Desktop/ML_Classification/')
```

#### 2.1 Getting the Data to Test

Loading the data as in step 0.1. There data as between 1 januar to 12 mars or so

```python
path="C:/Users/nelso/AppData/Roaming/MetaQuotes/Terminal/0F7D3A7524A217562675D753F7561259/tester/files/"
arr=os.listdir(path) #starting price array
data=[]
id=[]

for i in range(1,len(arr)):
    data.append(splt(arr[i]))
    id.append(path+arr[i])
    
print(len(id), "pictures")
```

#### 2.2 Testing the Data

Testing the data

```python
equity=100000
equity_array=[]
equity_array.append(100000)
print(len(id))
picturelist=[]
```

Trade Simulator

```python
# Infos for the classifier
batch_size = 32
img_height = 180
img_width = 180

# Infos on the class
class_names=['0','0.5','1','2','3','4']
num_classes = len(class_names)

#Creating an array for visualing the trade (going by open, close, result)
picturelist=[]

#Threshold 
threshold=2.5

#Comissions
comissions=0

for i in range(1,1100):
    #Classifier opening the picture
    img = tf.keras.utils.load_img(
    id[i], target_size=(img_height, img_width)
    )
    img_array = tf.keras.utils.img_to_array(img)
    img_array = tf.expand_dims(img_array, 0)
    predictions = model.predict(img_array)
    score = tf.nn.softmax(predictions[0])
    #Searching for the results
    if(float(class_names[np.argmax(score)])>threshold): #condition to open a trade
        risk=0.01*equity #risk (here is 1000.- each time, can change it to risk=0.01*equity_array[-1], last equity)
        if(float(data[i+1][0])<float(data[i][0])): #loss
            equity_array.append(equity_array[len(equity_array)-1]-risk)
        if(float(data[i+1][0])>float(data[i][0])): #win
            trade=risk*(float(data[i+1][3])-float(data[i+1][2]))/(float(data[i][3])-float(data[i][0]))-comissions
            res=equity_array[len(equity_array)-1]+trade #result
            equity_array.append(res)  
            print("trade taken : ",i,trade)
            if(trade>-2000): #modify to 0 if only positive trade or <500 for only negative trade
                picturelist.append((id[i],id[i+1],trade)) #add trade to picture list
```

#### 2.3 Plotting the result

Plotting the result of the equity

```python
plt.figure(figsize=(8,6))
plt.plot(equity_array,"-.",label=str(threshold) + "threshold")
plt.ylabel("Equity")
plt.xlabel("Trade Number")
plt.grid(color='0.6')
plt.title("Trading with the Classification Algo")
plt.legend()
```

Displaying picture_list trades

```python
for i in range(len(picturelist)):
    f, (ax1,ax2) = plt.subplots(1,2,sharey=True,figsize=((12,12)))
    ax1.axis('off')
    ax2.axis('off')
    ax1.imshow(np.asarray(Image.open(picturelist[i][0])))
    ax2.imshow(np.asarray(Image.open(picturelist[i][1])))
    f.text(0.5,0.3,"Trade result : "+str(picturelist[i][2]),ha='center')
```

```python
dest="C:/Users/nelso/desktop/ML_Classification/" #destination
for i in range(len(picturelist)):
    shutil.copyfile(picturelist[i][0],dest+str(i)+"_"+str(picturelist[i][2])+".gif")
    shutil.copyfile(picturelist[i][1],dest+str(i)+"_"+str(picturelist[i][2])+"_closed"+".gif")
print("Copying done")
```
